```python
import pandas as pd

df = pd.read_csv('data.csv')

df["Calories"].fillna(130, inplace = True)  # Substituir valores NULL apenas em colunas especificadas e pelo número 0

df
```

## Substituir células vazias usando média, mediana ou modo

```python
# .mean()

import pandas as pd

df = pd.read_csv('data.csv')

x = df["Calories"].mean()

df["Calories"].fillna(x, inplace = True)

df
```

```python
# .median()

import pandas as pd

df = pd.read_csv('data.csv')

x = df["Calories"].median()

df["Calories"].fillna(x, inplace = True)

df
```

```python
# .mode()

import pandas as pd

df = pd.read_csv('data.csv')

x = df["Calories"].mode()[0]

df["Calories"].fillna(x, inplace = True)

df
```

```python
# Converter em um formato correto

import pandas as pd

df = pd.read_csv('data.csv')

df['Date'] = pd.to_datetime(df['Date'])

print(df.to_string())

df
```

```python
df.dropna(subset=['Date'], inplace = True) # Campo data formatado
```

```python
# Remover dados duplicados

print(df.duplicated())      # Retorna Truepara cada linha duplicada, caso contrário False

df.drop_duplicates(inplace = True)      # Remova todas as duplicatas
```
