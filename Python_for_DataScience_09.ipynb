```python
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import rcParams

import seaborn as sns

# Configuração do Gráfico

%matplotlib inline
rcParams["figure.figsize"] = 5, 4
sns.set_style("whitegrid")
```

```python
x = range(1, 10)
y = [1, 2, 3, 4, 5, 4, 3, 2, 1]


plt.xlabel("Título do eixo X")
plt.ylabel("Título do eixo Y")

plt.bar(x, y)
```

```python
carros = pd.read_csv("mtcars.csv")
carros.columns = ["nome", "mpg", "cyl", "disp", "ªhp", "drat", "wt", "qsec", "vs", "am", "gear", "carb" ]

mpg = carros["mpg"]
```

```python
import matplotlib.patches as mpatches

plt.plot(mpg)
plt.xlabel("Nomes dos Carros")
plt.ylabel("Milhas/Galão")
plt.title("Consumo de Gasolina")

plt.xticks(mpg.index, carros.nome, rotation=45)

legenda = mpatches.Patch(label="mpg")
plt.legend(handles=[legenda])
```
